Cosine similarity measures the similarity between two vectors of an inner product space. It is measured by the cosine of the angle between two vectors and determines whether two vectors are pointing in roughly the same direction. It is often used to measure document similarity in text analysis.

```python
df=pd.DataFrame([{'Doc':doc1,'OEM':'microsoft'},{'Doc':doc2,'OEM':'tesla'},{'Doc':doc3,'OEM':'linux'},{'Doc':doc4,'OEM':'bmw'}])
df['Doc']=df['Doc'].apply(str.lower)
df
```

```python
df['Tokens']=df['Doc'].apply(word_tokenize)
df
```

```python
df['pText']=df['Tokens'].apply(lambda d:{w for w in d if not w in stopwords.words('english')})
df
```

```python
wLs=[]
for i in df['pText']:
  wLs.extend(list(i))
```

```python
# One Hot Encoding
def oneHotEnc(doc):
  ls=[]
  for w in wLs:
    if w in list(doc): ls.append(1)
    else: ls.append(0)
  return ls

df['OHE']=df['pText'].apply(oneHotEnc)
df
```

```python
# Cosine Similarity
def cosineSimilarity(r1,r2):
  c=0
  l1=df['OHE'][r1]
  l2=df['OHE'][r2]
  for i in range(len(wLs)):
          c+= l1[i]*l2[i]
  cosine = c / float((sum(l1)*sum(l2))**0.5)
  print(f"Cosine Similarity {df['OEM'][r1]} and {df['OEM'][r2]} DOCs is : {cosine}")

cosineSimilarity(0,1)
```

```python
from sklearn.metrics.pairwise import cosine_similarity, cosine_distances
cosine_similarity([df['OHE'][0]],[df['OHE'][1]])
```
