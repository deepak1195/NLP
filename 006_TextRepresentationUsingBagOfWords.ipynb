```python
from os import replace
df['spam']=df['Category'].apply(lambda x:1 if x=='spam' else 0)
df
```

```python
from sklearn.model_selection import train_test_split
xTrain,xTest,yTrain,yTest=train_test_split(df.Message, df.spam, test_size=0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vector=CountVectorizer()
xTrainCV=vector.fit_transform(xTrain)
xTrainCV
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(xTrainCV,yTrain)
```

```python
xTestCV=vector.transform(xTest)
```

```python
from sklearn.metrics import classification_report
yPred=model.predict(xTestCV)
```

```python
print(classification_report(yTest,yPred))
```

```python
from sklearn.pipeline import Pipeline

pModel=Pipeline([
    ('vectorizer', CountVectorizer()),
    ('NaivBais', MultinomialNB())
])

pModel.fit(xTrain,yTrain)
```

```python
yPred=pModel.predict(xTest)
print(classification_report(yTest,yPred))
```

```python
import plotly.express as px
import tensorflow as tf

cm=tf.math.confusion_matrix(labels=yTest,predictions=yPred)
px.imshow(
    cm,
    labels=dict(x="Predicted", y="Truth"),
    text_auto=True
)
```
